```python
## STANDARDIZE THE DATA
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.preprocessing import LabelEncoder
```

## Encoders

```python
sexLE = LabelEncoder()
```

```python
## sex encoder
X["worker_sex"]=sexLE.fit_transform(X["worker_sex"])
```

```python
X["chief_sex"] = sexLE.transform(X["chief_sex"])
```

```python
# Shift encoder
shiftLE = LabelEncoder()

X["worker_shift"]=shiftLE.fit_transform(X["worker_shift"])
```

```python
# Team encoder
teamLE = LabelEncoder()
X["worker_team"]=teamLE.fit_transform(X["worker_team"])
```

```python
# worker_role encoder
roleLE = LabelEncoder()
X["worker_role"]=roleLE.fit_transform(X["worker_role"])
```

```python
# chief_role encoder
chief_roleLE = LabelEncoder()
X["chief_role"]=chief_roleLE.fit_transform(X["chief_role"])
```

```python
# worker_workstyle encoder
worker_workstyleLE = LabelEncoder()
X["worker_workstyle"]=worker_workstyleLE.fit_transform(X["worker_workstyle"])
```

## Standard Scaler

```python
numerical_features = ['worker_age', 'chief_age', 'chief_sub_age_diff', 'worker_health',
                      'worker_commitment', 'worker_perceptiveness', 'worker_dexterity',
                      'worker_sociality', 'worker_goodness', 'worker_strength',
                      'worker_openmindedness', 'record_week_in_series', 'record_day_in_series']
```

```python
X_scaled = X.copy()

num_scalers=[]
cat_scalers=[]

for col in X.columns:
  if(col in numerical_features):
      scaler = StandardScaler()
      X_scaled[col] = scaler.fit_transform(X[col].values.reshape(-1, 1))
      num_scalers.append(scaler)
  else:
      scaler = MinMaxScaler()
      X_scaled[col] = scaler.fit_transform(X[col].values.reshape(-1, 1))
      cat_scalers.append(scaler)
X_scaled.head()

```
